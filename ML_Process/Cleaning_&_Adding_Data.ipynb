```python
import pandas as pd

NBA_df = pd.read_csv("../Resources/complete_data_ok.csv")

NBA_df=NBA_df.sort_values(by=['Player','Year'])
NBA_df=NBA_df.reset_index(drop=True)
NBA_df
```

```python
player=0
year=0
init=0
NBA_df["Played_Years"]=""
for index, row in NBA_df.iterrows():
    if(row['Player']!=player):
        player=row['Player']
        NBA_df.iloc[ index , 18 ]=init
        year=init
    else:
        year+=1
        NBA_df.iloc[ index , 18 ]=year
NBA_df
```
