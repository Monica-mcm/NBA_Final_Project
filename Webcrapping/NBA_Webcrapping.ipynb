```python
import pandas as pd
wiki_df = pd.read_html('https://en.wikipedia.org/wiki/NBA_Most_Valuable_Player_Award')
```

```python
wiki_df = wiki_df [5]
```

```python
wiki_df
```

```python
new_df_wiki = wiki_df[['Season', 'Player']].copy()
new_df_wiki
```

```python
new_df_wiki.dtypes
```

```python
new_df_wiki['Year'] = new_df_wiki['Season'].str[:2]+ new_df_wiki['Season'].str[5:]
new_df_wiki
```

```python
final_df_wiki = new_df_wiki [['Year', 'Player']].copy()
final_df_wiki
```

```python
import re
final_df_wiki['Player']
final_df_wiki
```

```python
def set_pandas_display_options() -> None:
    display = pd.options.display

    display.max_columns = 1000
    display.max_rows = 1000
    display.max_colwidth = 199
    display.width = None
    # display.precision = 2  # set as needed

set_pandas_display_options()
```

```python
for index, row in final_df_wiki.iterrows():
    final_df_wiki.loc[index, 'Players ok'] = re.sub("[^a-zA-Z ]", "", str(row['Player']))
final_df_wiki
```

```python
new_df_wiki = final_df_wiki [['Year', 'Players ok']].copy()
new_df_wiki
```

```python
new_df_wiki['Players ok'] = new_df_wiki['Players ok'].apply(lambda x: ' '.join(x.split(' ')[:2]))
new_df_wiki
```
