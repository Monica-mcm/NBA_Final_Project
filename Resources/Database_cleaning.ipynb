```python
# Loading players data into df
players = "players.csv"
players_df = pd.read_csv(players)
players_df
```

```python
# Loading seasons data into df
season_stats = "Seasons_Stats.csv"
season_stats_df = pd.read_csv(season_stats)
season_stats_df
```

```python
season_stats_df['Pos'].value_counts()
```

```python
# Printing seasons stats df columns
print(season_stats_df.columns)
```

```python
# Extracting only columns we need from season stats df and creating new df
new_season_stats_df = season_stats_df[['Year', 'Player', 'Pos', 'Age','G', 'GS', 'MP', 'PTS', 'PER','TS%', 'FTr', 'FG', 'FGA', 'AST', 'PF']].copy()
new_season_stats_df.rename(columns={'TS%':'TS'}, inplace=True)
new_season_stats_df.head()
```

```python
# Extracting only columns we need from season stats df and creating new df
new_players_df = players_df[['Player', 'height', 'weight']].copy()
new_players_df.head()
```

```python
# Turn Player names into strings for both df
new_season_stats_df['Player'] = new_season_stats_df['Player'].astype(str)
new_season_stats_df['Year'].fillna(0, inplace=True)
new_season_stats_df['Year'] = new_season_stats_df['Year'].astype(int)
new_players_df['Player'] = new_players_df['Player'].astype(str)
```

```python
# See type of variables in df's
new_season_stats_df.dtypes
```

```python
new_players_df.dtypes
```

```python
# Joining new_players_df onto new_season_stats_df 
complete_data_df = new_season_stats_df.merge(new_players_df, on='Player', how='inner')
```

```python
complete_data_df
```

```python
# Counting the amount of rows in new df
complete_data_df.count()
```

```python
# Delete * at the end of the Player's name
complete_data_df['Player'] = complete_data_df['Player'].apply(lambda x: x.replace('*',''))
complete_data_df
```

```python
# Doing another count as a check
complete_data_df.count()

# Export csv
complete_data_df.to_csv('complete_data_ok.csv')
```

```python
# MVP table 
mvp_players = "../Webcrapping/NBA_Webscrapping_ok.csv"
mvp_players_df = pd.read_csv(mvp_players)
mvp_players_df
```

```python
# Do left join on complete data df
df_with_mvp = complete_data_df.merge(mvp_players_df, how="left", left_on=['Year', 'Player'], right_on = ['Year', 'Players ok'])
df_with_mvp.head()
```

```python
# Use lambda function to replace Player's ok column NaN value with 0 and replace MVP winner with 1
df_with_mvp['Players ok'] = df_with_mvp['Players ok'].apply(lambda x: 1 if type(x) == str else 0)
```

```python
df_with_mvp.head()
```

```python
# Check if MVP title is assigned to the players
df_with_mvp_true = df_with_mvp[df_with_mvp['Players ok'] ==1]
df_with_mvp_true
```
