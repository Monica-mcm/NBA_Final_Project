```python
# Loading players data into df
players = "players.csv"
players_df = pd.read_csv(players)
players_df
```

```python
# Loading seasons data into df
season_stats = "Seasons_Stats.csv"
season_stats_df = pd.read_csv(season_stats)
season_stats_df
```

```python
# Printing seasons stats df columns
print(season_stats_df.columns)
```

```python
# Extracting only columns we need from season stats df and creating new df
new_season_stats_df = season_stats_df[['Year', 'Player', 'Pos', 'Tm', 'MP', 'PTS', 'PER']].copy()
new_season_stats_df.head()
```

```python
# Extracting only columns we need from season stats df and creating new df
new_players_df = players_df[['Player', 'born', 'height', 'weight', 'collage']].copy()
new_players_df.head()
```

```python
# Turn Player names into strings for both df
new_season_stats_df['Player'] = new_season_stats_df['Player'].astype(str)
new_players_df['Player'] = new_players_df['Player'].astype(str)
```

```python
# See type of variables in df's
new_season_stats_df.dtypes
```

```python
# Joining new_players_df onto new_season_stats_df 
complete_data_df = new_season_stats_df.merge(new_players_df, on='Player', how='inner')
```
